```python
from datasets import load_dataset

dataset = load_dataset("humarin/chatgpt-paraphrases")
dataset
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(dataset['train'][:])
df
```

```python
# i = df[((df.task == 'gec')|(df.task == 'clarity')|(df.task == 'simplification')|(df.task == 'coherence')|(df.task == 'neutralize'))].index
# df = df.drop(i).reset_index().drop(columns={'index','id','task'})
# df
#extracting only paraphrase task
i = df[((df.category== 'question'))].index
df = df.drop(i).reset_index()
df = df[['text','paraphrases']]
df.rename({'text': 'src','paraphrases':'tgt'}, axis=1, inplace=True)
new_tgt = []

df
```

```python
import ast

for i in df['tgt']:
    new_tgt.append(ast.literal_eval(i)[0])

df['tgt'] = new_tgt
df
```

```python
import re

def format_text(text):
    text = re.sub(r'([,.`!?])', r' \1 ', text)

    text = re.sub(r"\b(can|do|does|did|has|have|had|is|are|was|were|will|would|shall|should|must|might|need)n't\b", r"\1 n't", text)
    text = re.sub(r"(\w)('s|'ve|'re|'ll|'m|'d)\b", r"\1 \2", text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r"^'(?=\w)", r"' ", text)
    text = re.sub(r'^"(?=\w)', r'" ', text)
    return text
```

```python
formatted_src = []
formatted_tgt = []
for i in df['src']:
    formatted_src.append(format_text(i))
for j in df['tgt']:
    formatted_tgt.append(format_text(j))
df['src'] = formatted_src
df['tgt'] = formatted_tgt
df
```

```python
import numpy as np
def show_random_datapoints(n_samples,df):
    for i in range(n_samples):
        id = int(np.random.uniform(0,len(df)))

        if len(df['src'].iloc[id].split())>5:
            print(f"RESPONSE: {df['src'].iloc[id]}")
            print(f"SOLUTION: {df['tgt'].iloc[id]}")
            print('*'*100)
```

```python
show_random_datapoints(10,df)
```
